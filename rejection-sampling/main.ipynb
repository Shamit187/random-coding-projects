```python
import matplotlib.pyplot as plt
import numpy as np

# Rectangle bounds
left, right = -300, 300
bottom, top = -200, 200

# Generate random points
n_points = 2000
x_points = np.random.uniform(left, right, n_points)
y_points = np.random.uniform(bottom, top, n_points)

# Circle parameters
circle_radius = 100
circle_center = (0, 0)

# Plotting
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_xlim(left - 50, right + 50)
ax.set_ylim(bottom - 50, top + 50)

# Draw rectangle
rectangle = plt.Rectangle((left, bottom), right - left, top - bottom, fill=None, edgecolor='black', linewidth=2)
ax.add_patch(rectangle)

# Draw random points
ax.scatter(x_points, y_points, color='blue', s=10, label='Random Points')

# Draw circle
circle = plt.Circle(circle_center, circle_radius, color='red', fill=False, linewidth=2, label='Circle')
ax.add_patch(circle)

# Formatting
ax.set_aspect('equal')
plt.grid(True)
plt.savefig('random_points_in_circle.png', dpi=300)
```
